```python
# Крестики-нолики (консольная версия)

def print_board(board):
    print("  0 1 2")
    for i, row in enumerate(board):
        print(i, " ".join(row))


def check_win(board, player):
    # строки
    for row in board:
        if all(cell == player for cell in row):
            return True

    # столбцы
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True

    # диагонали
    if all(board[i][i] == player for i in range(3)):
        return True

    if all(board[i][2 - i] == player for i in range(3)):
        return True

    return False


def is_full(board):
    return all(cell != "-" for row in board for cell in row)


def get_move(player):
    while True:
        try:
            row = int(input(f"Игрок {player}, введите строку (0-2): "))
            col = int(input(f"Игрок {player}, введите столбец (0-2): "))

            if row not in range(3) or col not in range(3):
                print("❌ Координаты должны быть от 0 до 2")
                continue

            return row, col
        except ValueError:
            print("❌ Введите числа!")


def main():
    board = [["-" for _ in range(3)] for _ in range(3)]
    player = "x"

    print("Игра КРЕСТИКИ-НОЛИКИ начинается!\n")

    while True:
        print_board(board)

        row, col = get_move(player)

        if board[row][col] != "-":
            print("❌ Клетка занята, попробуйте ещё раз")
            continue

        board[row][col] = player

        if check_win(board, player):
            print_board(board)
            print(f"🎉 Игрок {player} победил!")
            break

        if is_full(board):
            print_board(board)
            print("🤝 Ничья!")
            break

        player = "o" if player == "x" else "x"


if __name__ == "__main__":
    main()
```
